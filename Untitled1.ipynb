```python
import pandas as pd

# Load
df = pd.read_csv('dataset.csv')

# Timestamp
df['Timestamp'] = pd.to_datetime(df['LastUpdatedDate'] + ' ' + df['LastUpdatedTime'],
                                 format='%d-%m-%Y %H:%M:%S')

# Clean traffic
df['TrafficConditionNearby'] = df['TrafficConditionNearby'].astype(str).str.strip().str.lower()
traffic_map = {'low': 1.0, 'average': 2.0, 'high': 3.0}
df['TrafficConditionNearby'] = df['TrafficConditionNearby'].map(traffic_map)

# Drop missing rows
df = df.dropna()

# Save for Pathway
df[["Timestamp", "Occupancy", "Capacity", "QueueLength",
    "TrafficConditionNearby", "VehicleType", "IsSpecialDay",
    "Latitude", "Longitude"]].to_csv("parking_stream.csv", index=False)
```

 STEP 4: Setup Pathway Streaming

```python
import pathway as pw

class ParkingSchema(pw.Schema):
    Timestamp: str
    Occupancy: int
    Capacity: int
    QueueLength: int
    TrafficConditionNearby: float
    VehicleType: str
    IsSpecialDay: int
    Latitude: float
    Longitude: float

# Stream data
data = pw.demo.replay_csv("parking_stream.csv", schema=ParkingSchema, input_rate=100)

# Time parsing
fmt = "%Y-%m-%d %H:%M:%S"
data_with_time = data.with_columns(
    t = data.Timestamp.dt.strptime(fmt),
    day = data.Timestamp.dt.strptime(fmt).dt.strftime("%Y-%m-%dT00:00:00")
)
```

MODEL 1: BASELINE LINEAR

```python
import datetime
base_price = 10
alpha = 5

model1 = (
    data_with_time.windowby(
        pw.this.t,
        instance=pw.this.day,
        window=pw.temporal.tumbling(datetime.timedelta(days=1)),
        behavior=pw.temporal.exactly_once_behavior()
    ).reduce(
        t = pw.this._pw_window_end,
        occ_sum = pw.reducers.sum(pw.this.Occupancy),
        cap_sum = pw.reducers.sum(pw.this.Capacity)
    ).with_columns(
        occ_rate = pw.this.occ_sum / pw.this.cap_sum,
        price = base_price + alpha * (pw.this.occ_sum / pw.this.cap_sum)
    )
)
```

 MODEL 2: DEMAND-BASED

```python
MIN_DEMAND = 0
MAX_DEMAND = 10

reduced = (
    data_with_time.windowby(
        pw.this.t,
        instance=pw.this.day,
        window=pw.temporal.tumbling(datetime.timedelta(days=1)),
        behavior=pw.temporal.exactly_once_behavior()
    ).reduce(
        t = pw.this._pw_window_end,
        occ_sum = pw.reducers.sum(pw.this.Occupancy),
        cap_sum = pw.reducers.sum(pw.this.Capacity),
        queue_sum = pw.reducers.sum(pw.this.QueueLength),
        traffic_sum = pw.reducers.sum(pw.this.TrafficConditionNearby),
        special_day = pw.reducers.max(pw.this.IsSpecialDay),
        vehicle_weight_sum = pw.reducers.sum(
            pw.if_else(pw.this.VehicleType == "bike", 0.5,
                       pw.if_else(pw.this.VehicleType == "truck", 1.5, 1.0))
        ),
        row_count = pw.reducers.count()
    )
)

averaged = reduced.with_columns(
    t = pw.this.t,
    occ_rate_avg = pw.this.occ_sum / pw.this.cap_sum,
    queue_avg = pw.this.queue_sum / pw.this.row_count,
    traffic_avg = pw.this.traffic_sum / pw.this.row_count,
    vehicle_score = pw.this.vehicle_weight_sum / pw.this.row_count
)

with_demand = averaged.with_columns(
    t = pw.this.t,
    demand = (
        1.0 * pw.this.occ_rate_avg +
        0.8 * pw.this.queue_avg -
        0.5 * pw.this.traffic_avg +
        1.2 * pw.this.special_day +
        1.5 * pw.this.vehicle_score
    )
)

with_norm = with_demand.with_columns(
    t = pw.this.t,
    norm_demand = (pw.this.demand - MIN_DEMAND) / (MAX_DEMAND - MIN_DEMAND + 1e-5)
)

model2 = with_norm.with_columns(
    t = pw.this.t,
    price = 10 + 10 * pw.this.norm_demand  # base * (1 + lambda*demand)
)
```

MODEL 3: OPTIONAL COMPETITIVE (SIMULATED)

```python
model3 = with_norm.with_columns(
    t = pw.this.t,
    price = pw.if_else(
        pw.this.occ_rate_avg > 0.9,
        0.8 * (10 + 10 * pw.this.norm_demand),  # reduce price if nearly full
        1.1 * (10 + 10 * pw.this.norm_demand)   # otherwise increase slightly
    )
)
```

EXPORT & PLOT MODELS

```python
import tempfile, json, matplotlib.pyplot as plt

def export_and_plot(model, label, color):
    path = tempfile.NamedTemporaryFile(delete=False).name
    pw.io.jsonlines.write(model, path)
    pw.run()
    with open(path, "r") as f:
        lines = [json.loads(line) for line in f]
    df = pd.DataFrame(lines)
    df['t'] = pd.to_datetime(df['t'])
    df = df.sort_values('t')
    plt.plot(df['t'], df['price'], label=label, marker='o', linewidth=2, color=color)

plt.figure(figsize=(12, 6))
export_and_plot(model1, "Model 1: Linear", "green")
export_and_plot(model2, "Model 2: Demand-Based", "blue")
export_and_plot(model3, "Model 3: Competitive", "red")
plt.title("Dynamic Parking Price Models Over Time")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
